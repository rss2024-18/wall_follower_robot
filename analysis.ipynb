```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
subfolder = 'charles'

test_cases = ['short_right_close', 'short_left_far', 'short_right_angled', 'short_left_far_angled', 'long_right', 'long_left']
for test in test_cases:
    file_distance = np.load('./test_data/' + subfolder + '/' + test + '_log.npz')
    file_steering = np.load('./test_data/' + subfolder + '/' + test + '_steer.npz')
    distances = file_distance['arr_0']
    steering = file_steering['arr_0']

    plt.scatter(distances[:,0], distances[:,3], s=3)
    plt.title(test + " distances")
    plt.xlabel('Time')
    plt.ylabel('Distance')
    # plt.ylim(0.5, 1.5)

    plt.savefig('./test_data/' + subfolder + '/' + test + '_distances.png')
    plt.figure()

    plt.scatter(steering[:,0], steering[:,1], s=3)
    plt.title(test + " steering")
    plt.xlabel('Time')
    plt.ylabel('Angle Setpoint')
    # plt.ylim(-np.pi, np.pi)

    plt.savefig('./test_data/' + subfolder + '/' + test + '_steering.png')
    plt.figure()
```
